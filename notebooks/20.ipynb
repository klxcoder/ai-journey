# Predictive coding

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.typing import NDArray

# Generate data
np.random.seed(0)
x: NDArray[np.float64] = np.linspace(-1, 1, 100).reshape(-1, 1)
y_true: NDArray[np.float64] = 2 * x + 1 + 0.1 * np.random.randn(*x.shape)

# Initialize weights and bias
w: float = np.random.randn()
b: float = np.random.randn()

# Predictive coding loop
lr: float = 0.1
epochs: int = 50

for epoch in range(epochs):
    # Forward prediction
    y_pred = w * x + b
    
    # Prediction error
    error = y_true - y_pred
    
    # Update parameters to minimize error (local update rules)
    w += lr * float(np.mean(error * x))
    b += lr * float(np.mean(error))

# Plot
plt.scatter(x, y_true, label='True')
plt.plot(x, w * x + b, color='red', label='Predicted')
plt.title('Predictive Coding Linear Regression')
plt.legend()
plt.show()
```
